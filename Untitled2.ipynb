```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Полный анализ распределения слов
def analyze_tokenized_text(df, text_column='lemmatized_text', top_n=20):
    """
    Полный анализ распределения слов в токенизированном тексте
    """
    # Собираем все слова в один список
    all_words = []
    for sublist in df[text_column]:
        if isinstance(sublist, list):  # Проверяем, что это список
            all_words.extend(sublist)

    # Подсчет частот
    word_freq = Counter(all_words)

    # Создаем DataFrame с распределением
    freq_df = pd.DataFrame(word_freq.most_common(),
                          columns=['word', 'frequency'])

    # Добавляем проценты
    total_tokens = len(all_words)
    freq_df['percentage'] = (freq_df['frequency'] / total_tokens) * 100
    freq_df['cumulative_percentage'] = freq_df['percentage'].cumsum()

    # Базовая статистика
    stats = {
        'total_tokens': total_tokens,
        'unique_words': len(freq_df),
        'avg_word_frequency': freq_df['frequency'].mean(),
        'median_word_frequency': freq_df['frequency'].median(),
        'most_common_word': freq_df.iloc[0]['word'],
        'most_common_frequency': freq_df.iloc[0]['frequency']
    }

    return freq_df, stats, all_words

# Запускаем анализ
freq_df, stats, all_words = analyze_tokenized_text(df)

# Выводим результаты
print("=" * 50)
print("СТАТИСТИКА РАСПРЕДЕЛЕНИЯ СЛОВ")
print("=" * 50)
print(f"Общее количество токенов: {stats['total_tokens']:,}")
print(f"Количество уникальных слов: {stats['unique_words']:,}")
print(f"Средняя частота слова: {stats['avg_word_frequency']:.2f}")
print(f"Медианная частота: {stats['median_word_frequency']:.2f}")
print(f"Самое частое слово: '{stats['most_common_word']}' ({stats['most_common_frequency']} раз)")
print(f"Размер словаря: {stats['unique_words']:,} слов")

# Визуализация
plt.figure(figsize=(15, 10))


# График 4: Длины слов
plt.subplot(2, 2, 4)
word_lengths = [len(str(word)) for word in all_words]
length_freq = Counter(word_lengths)
length_df = pd.DataFrame(length_freq.items(), columns=['length', 'count']).sort_values('length')
sns.barplot(data=length_df, x='length', y='count', palette='coolwarm')
plt.title('Распределение длин слов')
plt.xlabel('Длина слова')
plt.ylabel('Количество')

plt.tight_layout()
plt.show()

# Дополнительная информация
print("\n" + "=" * 50)
print("ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ")
print("=" * 50)

# Слова, которые встречаются только один раз (гапакс legomena)
hapax_words = freq_df[freq_df['frequency'] == 1]
print(f"Слов, встречающихся только 1 раз: {len(hapax_words):,}")
print(f"Это {len(hapax_words)/stats['unique_words']*100:.1f}% от всего словаря")

# Сколько слов покрывают 50%, 80%, 90% текста
for coverage in [50, 80, 90, 95]:
    coverage_idx = freq_df[freq_df['cumulative_percentage'] >= coverage].index[0]
    print(f"Для покрытия {coverage}% текста нужно {coverage_idx + 1} самых частых слов")

# Статистика по длинам слов
print(f"\nСредняя длина слова: {np.mean(word_lengths):.2f} символов")
print(f"Медианная длина слова: {np.median(word_lengths):.2f} символов")
print(f"Самая частая длина слова: {length_df.loc[length_df['count'].idxmax(), 'length']} символов")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Убедись, что это строки (ты уже проверил — это так)
# df['lemmatized_text'] = "получен куар ход" и т.д.

# Создаём векторизатор
vectorizer = TfidfVectorizer(
    lowercase=False,        # не менять регистр (у тебя уже леммы)
    token_pattern=r'\S+'    # токены = непробельные последовательности
)

# Применяем напрямую к строкам
tfidf_matrix = vectorizer.fit_transform(df['lemmatized_text'])

# Получаем длинный формат: word | coord | values
doc_indices, word_indices = tfidf_matrix.nonzero()
tfidf_values = tfidf_matrix.data
feature_names = vectorizer.get_feature_names_out()

tfidf_long = pd.DataFrame({
    'coord': doc_indices,
    'word': feature_names[word_indices],
    'values': tfidf_values
}).sort_values(['coord', 'values'], ascending=[True, False]).reset_index(drop=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Убедимся, что у нас есть vectorizer и tfidf_matrix
# (если нет — сначала выполните код из предыдущего ответа)

# Среднее TF-IDF по каждому слову
mean_tfidf = np.array(tfidf_matrix.mean(axis=0)).flatten()
feature_names = vectorizer.get_feature_names_out()

# Создаём DataFrame для удобства
tfidf_word_avg = pd.DataFrame({
    'word': feature_names,
    'mean_tfidf': mean_tfidf
}).sort_values('mean_tfidf', ascending=False)

# Берём топ-20
top_n = 20
top_words = tfidf_word_avg.head(top_n)

# График
plt.figure(figsize=(10, 8))
plt.barh(top_words['word'][::-1], top_words['mean_tfidf'][::-1], color='skyblue')
plt.xlabel('Среднее TF-IDF значение')
plt.title(f'Топ-{top_n} самых значимых слов (по среднему TF-IDF)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
